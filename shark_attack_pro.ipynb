```python
def standardize_time_group (Time):
  if pd.isna(Time):
    return 'unknown'

  Time = str(Time).lower().strip()  # Normalize for comparison

  if 'morning' in Time or 'am' in Time:
      return 'morning'
  if 'afternoon' in Time or 'pm' in Time or 'midday' in Time:
      return 'afternoon'
  if 'dusk' in Time or 'evening' in Time:
      return 'evening'
  if 'night' in Time:
      return 'night'
  if ':' in Time or 'h' in Time:
      return 'exact'

  return 'unknown'

df_shark['Time'] = df_shark['Time'].apply(standardize_time_group)
```

```python
#Grouping the name of the Activity rows values
def standardize_activity(Activity):
    if pd.isna(Activity):
        return 'unknown'
    if 'surf' in Activity:
        return 'surfing'
    elif 'swim' in Activity:
        return 'swimming'
    elif 'diving' in Activity or 'dive' in Activity:
        return 'diving'
    elif 'fish' in Activity:
        return 'fishing'
    elif 'kayak' in Activity or 'canoe' in Activity:
        return 'kayaking'
    elif 'stand-up paddle' in Activity or 'paddle' in Activity:
        return 'paddleboarding'
    elif 'snorkel' in Activity:
        return 'snorkeling'
    elif 'boat' in Activity or 'sail' in Activity:
        return 'boating'
    elif 'wading' in Activity:
        return 'wading'
    elif 'walking' in Activity or 'standing' in Activity:
        return 'walking/standing'
    else:
        return 'other'
```

```python
def clean_age(Age):
       # Convert to string and lowercase
    Age = str(Age).lower().strip()
    # Direct numbers
    if Age.isdigit():
        return float(Age)
    # Extract numbers like "35 years", "~27", "30s", "18 months", "20-25"
    match = re.search(r'\d{1,3}', Age) #regex
    if match:
        return float(match.group())
    # Handle text-based age categories
    if 'teen' in Age:
        return 15.0
    elif 'month'in Age:
        return 1.0
    elif 'baby' in Age or 'infant' in Age:
        return 1.0
    elif 'toddler' in Age:
        return 2.0
    elif 'child' in Age:
        return 8.0
    elif 'middle' in Age:
        return 45.0
    elif 'adult' in Age:
        return 30.0
    elif 'young' in Age:
        return 20.0
    elif 'elderly' in Age or 'senior' in Age:
        return 70.0
    elif '?' in Age:
        return None
    return None  # fallback if we can’t parse it

```

```python
def clean_dataset(df: pd.DataFrame) -> pd.DataFrame:

  df2 = df.copy()

  #Drop unnecessary column
  df2 = df2.drop(columns=['Source', 'pdf', 'href formula', 'href', 'Case Number', 'Case Number.1',
       'original order', 'Unnamed: 21', 'Unnamed: 22'])

  # Remove leading/trailing white spaces from column names
  df2.columns = df2.columns.str.strip()

  # Delete 'Y/N' from 'Fatal ', deleting part of a column name
  df2.columns = df2.columns.str.replace('Y/N', '')

  df2['Sex'] = df2['Sex'].fillna('unknown').replace({'F ': 'F',
                                                     'M ': 'M' ,
                                                     ' M': 'M' ,
                                                     'lli': 'unknown',
                                                     'M x 2': 'M',
                                                     'N': 'unknown',
                                                     '.': 'unknown'})

  df2['Sex'] = df2['Sex'].replace(np.nan, 'unknown')

  #Activitiy function calling
  df2['Activity'] = df2['Activity'].apply(standardize_activity)

  #Lowercase all the rows in standart
  df2['Activity']=df2['Activity'].str.lower().str.strip()

  # Fixing "age"
  df2['Age'] = pd.to_numeric(df2['Age'], errors='coerce')
  df2 = df2.dropna(subset=['Age'])

  df2['Age'] = df2['Age'].apply(clean_age)
  #df2.rename(columns= {'age_cleaned': 'Age'}, inplace=True)

  #Location fillna
  df_shark['Location']=df_shark['Location'].fillna('unknown')

  #Dropping nan values for Country
  df2=df2.dropna(subset=['Country'])

  #Fillna unknown Injury nan values
  df2['Injury']=df2['Injury'].fillna('unknown')

  # Remove leading/trailing white spaces from column names
  df2.columns = df2.columns.str.strip()

  #Standardized Fatal row values
  # Step 1: Remove leading/trailing spaces and make uppercase
  df2['Fatal'] = df2['Fatal'].str.strip().str.upper()

  valid_values = ['Y', 'N']

  df2['Fatal'] = df2['Fatal'].apply(lambda x: x if x in valid_values else 'unknown')

  #DateTime format standardized
  df2['Date']=pd.to_datetime(df2['Date'], errors ='coerce')
  #If it's already datetime (with NaT):
  df2['Date'] = df2['Date'].fillna(pd.Timestamp('1900-01-01'))  # a clear placeholder

  #Year format numerical only
  df2['Year'] = pd.to_numeric(df2['Year'], errors='coerce')

  #Standardized Type row values
  valid_value_type= ['Unprovoked', 'Questionable', 'Provoked', 'Watercraft', 'Sea Disaster','Under investigation']

  df_shark['Type']=df_shark['Type'].apply(lambda x: x if x in valid_value_type else 'unknown')

  return df2
```

```python
clean_dataset(df_shark)
df_shark= clean_dataset(df_shark)
```
